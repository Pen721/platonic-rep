```python
import requests
import os

urls = [
    "https://data.statmt.org/opus-100-corpus/v1.0/supervised/en-zh/opus.en-zh-test.zh",
    "https://data.statmt.org/opus-100-corpus/v1.0/supervised/en-zh/opus.en-zh-test.en"
]

def get_data(url):
    filename = url.split('/')[-1]
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            return f.read().splitlines()
    else:
        response = requests.get(url)
        with open(filename, 'w') as f:
            f.write(response.text)
        return response.text.splitlines()

zh_data = get_data(urls[0])
en_data = get_data(urls[1])
```

```python
for x, y in zip(zh_data, en_data):
    print(x)
    print(y)
    print()

```
